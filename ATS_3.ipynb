LAB 3A



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
np.random.seed(42)
n = 100
# -----------------------------
# 1. Create Different Series
# -----------------------------
# White noise (stationary)
wn = np.random.normal(0,1,n)
series_wn = pd.Series(wn, name='White Noise')
# Oscillatory (stationary with seasonality)
osc = np.sin(2*np.pi*np.arange(n)/10) + np.random.normal(0,0.3,n)
series_osc = pd.Series(osc, name='Oscillatory')
# Trending series (non-stationary)
trend = np.arange(n) + np.random.normal(0,1,n)
series_trend = pd.Series(trend, name='Trend')
# -----------------------------
# 2. Plot All Series
# -----------------------------
plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.plot(series_wn); plt.title('White Noise Series')
plt.subplot(3,1,2)
plt.plot(series_osc); plt.title('Oscillatory Series')
plt.subplot(3,1,3)
plt.plot(series_trend); plt.title('Trending Series')
plt.tight_layout(); plt.show()
# -----------------------------
# 3. Compute & Print ACF
# -----------------------------
def print_acf(series, name):
    acf_vals = acf(series, nlags=20)
    print(f"\nACF Values for {name}:")
    for i, val in enumerate(acf_vals):
        print(f"Lag {i}: {val:.4f}")
print_acf(series_wn, "White Noise")
print_acf(series_osc, "Oscillatory")
print_acf(series_trend, "Trend")
# -----------------------------
# 4. Plot ACF
# -----------------------------
plt.figure(figsize=(12,8))
plt.subplot(3,1,1); plot_acf(series_wn, lags=20, ax=plt.gca()); plt.title('ACF: White Noise')
plt.subplot(3,1,2); plot_acf(series_osc, lags=20, ax=plt.gca()); plt.title('ACF: Oscillatory')
plt.subplot(3,1,3); plot_acf(series_trend, lags=20, ax=plt.gca()); plt.title('ACF: Trend')
plt.tight_layout(); plt.show()
```

LAB 3B

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, adfuller
from statsmodels.graphics.tsaplots import plot_acf
np.random.seed(42)
n = 100
# -----------------------------
# 1. Trending Series
# -----------------------------
trend = np.arange(n) + np.random.normal(0,1,n)
series_trend = pd.Series(trend, name='Trend')
# -----------------------------
# 2. Original Series: Mean & Std
# -----------------------------
mean_orig = series_trend.mean()
std_orig = series_trend.std()
print(f"Original Series Mean: {mean_orig:.4f}, Std: {std_orig:.4f}")
# -----------------------------
# 3. Original Series: ACF
# -----------------------------
acf_orig = acf(series_trend, nlags=20)
print("\nACF Values of Original Series:")
for i, val in enumerate(acf_orig):
 print(f"Lag {i}: {val:.4f}")
# -----------------------------
# 4. Original Series: ADF Test
# -----------------------------
adf_orig = adfuller(series_trend)
print(f"\nADF Statistic (Original): {adf_orig[0]:.4f}, p-value: {adf_orig[1]:.4f}")
if adf_orig[1] < 0.05:
 print("Original series is stationary")
else:
 print("Original series is non-stationary")
# -----------------------------
# 5. First-Order Differencing
# -----------------------------
diff_series = series_trend.diff().dropna()
# -----------------------------
# 6. Differenced Series: Mean & Std
# -----------------------------
mean_diff = diff_series.mean()
std_diff = diff_series.std()
print(f"\nDifferenced Series Mean: {mean_diff:.4f}, Std: {std_diff:.4f}")
# -----------------------------
# 7. Differenced Series: ACF
# -----------------------------
acf_diff = acf(diff_series, nlags=20)
print("\nACF Values of Differenced Series:")
for i, val in enumerate(acf_diff):
 print(f"Lag {i}: {val:.4f}")
# -----------------------------
# 8. Differenced Series: ADF Test
# -----------------------------
adf_diff = adfuller(diff_series)
print(f"\nADF Statistic (Differenced): {adf_diff[0]:.4f}, p-value: {adf_diff[1]:.4f}")
if adf_diff[1] < 0.05:
 print("Differenced series is stationary")
else:
 print("Differenced series is non-stationary")
# -----------------------------
# 9. Plot Series and ACF
# -----------------------------
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.plot(series_trend); plt.title('Original Series')
plt.subplot(2,2,2)
plot_acf(series_trend, lags=20, ax=plt.gca()); plt.title('ACF: Original Series')
plt.subplot(2,2,3)
plt.plot(diff_series); plt.title('Differenced Series')
plt.subplot(2,2,4)
plot_acf(diff_series, lags=20, ax=plt.gca()); plt.title('ACF: Differenced Series')
plt.tight_layout(); plt.show()
```
